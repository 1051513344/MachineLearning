```python
# 0.
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 5) for i in x]

# 1.
# plt.figure(figsize=(20, 8), dpi=100)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=100)


# 2.
# plt.plot(x, y_shanghai, label="")
# plt.plot(x, y_beijing, color="r", linestyle="--", label="")
axes[0].plot(x, y_shanghai, label="")
axes[1].plot(x, y_beijing, color="r", linestyle="--", label="")

# 2.1 x,y
# x,y
x_ticks_label = ["11{}".format(i) for i in x]
y_ticks = range(40)

# 
# plt.xticks(x[::5], x_ticks_label[::5])
# plt.yticks(y_ticks[::5])
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks[::5])
axes[0].set_xticklabels(x_ticks_label[::5], fontproperties=font)
axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks[::5])
axes[1].set_xticklabels(x_ticks_label[::5])

    
    
# 2.2 
# plt.grid(True, linestyle="--", alpha=0.5)
axes[0].grid(True, linestyle="--", alpha=0.5)
axes[1].grid(True, linestyle="--", alpha=0.5)

# 2.3 
# plt.xlabel("")
# plt.ylabel("")
# plt.title("11--12", fontsize=20)
axes[0].set_xlabel("", fontproperties=font)
axes[0].set_ylabel("", fontproperties=font)
axes[0].set_title("11--12", fontsize=20, fontproperties=font)
axes[1].set_xlabel("", fontproperties=font)
axes[1].set_ylabel("", fontproperties=font)
axes[1].set_title("11--12", fontsize=20, fontproperties=font)

# # 2.4 
# plt.savefig("./test.png")

# # 2.5 
# plt.legend(loc=0)
axes[0].legend(loc=0, prop=font)
axes[1].legend(loc=0, prop=font)


# 3.
plt.show()
```
