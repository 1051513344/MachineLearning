# 1.2 添加自定义x,y刻度

```python
# 增加以下两行代码


# 画出温度变化图

# 0.准备x, y坐标的数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]

# 1.创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 2.绘制折线图
plt.plot(x, y_shanghai)

# 构造x轴刻度标签
x_ticks_label = ["11点{}分".format(i) for i in x]
# 构造y轴刻度
y_ticks = range(40)


# 修改x,y轴坐标的刻度显示
_x = plt.xticks(x[::5], x_ticks_label[::5])
_y = plt.yticks(y_ticks[::5])
for f in _x[1]:
    f.set_fontproperties(font)
# 网格
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel("时间", fontproperties=font)
plt.ylabel("温度", fontproperties=font)
plt.title("中午11点0分到12点之间的温度变化图示", fontsize=20, fontproperties=font)
# 保存图片到指定路径
plt.savefig("test.png")
# 3.显示图像
plt.show()

```

```python
# 0.准备数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 5) for i in x]

# 1.创建画布
# plt.figure(figsize=(20, 8), dpi=100)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=100)


# 2.绘制图像
# plt.plot(x, y_shanghai, label="上海")
# plt.plot(x, y_beijing, color="r", linestyle="--", label="北京")
axes[0].plot(x, y_shanghai, label="上海")
axes[1].plot(x, y_beijing, color="r", linestyle="--", label="北京")

# 2.1 添加x,y轴刻度
# 构造x,y轴刻度标签
x_ticks_label = ["11点{}分".format(i) for i in x]
y_ticks = range(40)

# 刻度显示
# plt.xticks(x[::5], x_ticks_label[::5])
# plt.yticks(y_ticks[::5])
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks[::5])
axes[0].set_xticklabels(x_ticks_label[::5], fontproperties=font)
axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks[::5])
axes[1].set_xticklabels(x_ticks_label[::5])

    
    
# 2.2 添加网格显示
# plt.grid(True, linestyle="--", alpha=0.5)
axes[0].grid(True, linestyle="--", alpha=0.5)
axes[1].grid(True, linestyle="--", alpha=0.5)

# 2.3 添加描述信息
# plt.xlabel("时间")
# plt.ylabel("温度")
# plt.title("中午11点--12点某城市温度变化图", fontsize=20)
axes[0].set_xlabel("时间", fontproperties=font)
axes[0].set_ylabel("温度", fontproperties=font)
axes[0].set_title("中午11点--12点某城市温度变化图", fontsize=20, fontproperties=font)
axes[1].set_xlabel("时间", fontproperties=font)
axes[1].set_ylabel("温度", fontproperties=font)
axes[1].set_title("中午11点--12点某城市温度变化图", fontsize=20, fontproperties=font)

# # 2.4 图像保存
# plt.savefig("./test.png")

# # 2.5 添加图例
# plt.legend(loc=0)
axes[0].legend(loc=0, prop=font)
axes[1].legend(loc=0, prop=font)


# 3.图像显示
plt.show()
```
